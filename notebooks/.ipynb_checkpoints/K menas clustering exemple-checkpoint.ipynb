## Method Used

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

```python
# a primeira array tem 200 samples e 2 features conforme selecionamos em cima 
data[0].shape
```

```python
# plotar todas as linhas da primeira coluna  do array e todas linhas da segunda coluna
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow') # c = bloobs, grupos
```

```python
# bloobs, marcador dos grupos de cima, grupo0,1,2 ou 3 
# em um caso real n√£o ter√≠amos esse valor, eles que gostar√≠amos de saber
data[1]
```

Creating the clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
# centro de cada grupo que o algor√≠timo acredita ser
kmeans.cluster_centers_
```

```python
# nova array com os grupos definidos pelo algor√≠timo
# em um projeto real n√£o ter√≠amos uma array pr√© definida com o n√∫mero de grupos para comparar
# ent√£o o n_clusters=4 seria um chute, pois n√£o saber√≠amos quantos grupos poderiam se formar ali
# deveriamos ent√£o tentar v√°rios valores de clusters at√© chegar a um padr√£o aceit√°vel
kmeans.labels_
```

```python
# cria√ß√£o de subplots de 1 linha e 2 colunas,usam o mesmo x
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6),)

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')

```
